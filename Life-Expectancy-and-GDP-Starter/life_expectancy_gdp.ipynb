```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('all_data.csv')
df.head()
```

```python
# life expectancy of each country over time
sns.set_style('darkgrid')
sns.set_palette('pastel')
f, ax = plt.subplots(figsize=(12,8))
sns.lineplot(data=df, x='Year', y='Life expectancy at birth (years)', hue='Country')
plt.show()
plt.clf()
```

```python
# gdp of each country over time
f, ax = plt.subplots(figsize=(12, 8))
sns.lineplot(data=df, x='Year', y='GDP', hue='Country')
plt.show()
plt.clf()
```

```python
# life expectancy and gdp correlation for each country
countries = df.Country.unique()
df['GDP in Trillion'] = df.GDP / 1000000000000
f = plt.figure(figsize=(15, 15))
ax = plt.subplot(3,2,6)
for i, country in zip(range(1, 7), countries):
    axi = plt.subplot(3,2,i)
    sns.scatterplot(data=df[df.Country == country], x='GDP in Trillion', y='Life expectancy at birth (years)')
    plt.title('Life Expectancy vs GDP for {}'.format(country))
#     axi.set_xticks('')
    
plt.subplots_adjust(wspace=.3, hspace=.4)
plt.show()
plt.clf()
```

```python
# averge life expectancy for each country
f, ax = plt.subplots(figsize=(12, 8))
sns.barplot(x='Country', y='Life expectancy at birth (years)', data=df)
plt.show()
plt.clf()
```

```python
# distribution of life expectancy for each country
f, ax = plt.subplots(figsize=(12, 8))
for country in countries:
    sns.kdeplot(df['Life expectancy at birth (years)'][df.Country == country], shade=True, label=country)
plt.legend(loc=2)
plt.show()
plt.clf()
```

### Conclusion

- life expectancy increasing over time
- GDP increasing over time
- life expectancy positively correlated to GDP
- Zimbabwe's life expectancy range is very large compared to the other countries
