```python
# create 'genre' column
def parse_values(x):
    if x == 'League of Legends' or x == 'Dota 2' or x == 'Heroes of the Storm':
        return 'MOBA'
    elif x == 'Counter-Strike: Global Offensive':
        return 'FPS'
    elif x == 'DayZ' or x == 'ARK: Survival Evolved':
        return 'Survival'
    else:
        return 'Other'
stream['genre'] = stream.game.apply(parse_values)
```

```python
# create 'hour' and 'day_of_week' columns to compare views for particular day and time
stream.time = pd.to_datetime(stream.time, format='%Y-%m-%d %H:%M:%S')
stream['hour'] = stream.time.dt.hour
stream['day_of_week'] = stream.time.dt.dayofweek
pd.crosstab(stream.hour, stream.day_of_week)
```
