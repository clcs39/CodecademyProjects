```python
# update damages data
for a, i in enumerate(damages):
    if i[-1] == 'M':
        damages[a] = float(i.strip('M')) * 1000000
        
    elif i[-1] == 'B':
        damages[a] = float(i.strip('B')) * 1000000000
    else:
        continue
damages
```

```python
# create hurricanes dictionary
dct = {}
for a,b,c,d,e,f,g in zip(names, months, years, max_sustained_winds, areas_affected, damages, deaths):
    sub_dct = {}
    sub_dct['Name'] = a
    sub_dct['Month'] = b
    sub_dct['Year'] = c
    sub_dct['Max Sustained Wind'] = d
    sub_dct['Areas Affected'] = e
    sub_dct['Damage'] = f
    sub_dct['Deaths'] = g
    dct[a] = sub_dct
print(dct)
```

```python
# create a new dictionary of hurricanes with year and key
def find_hurricane(year):
    dct1 = {}
    lst = []
    for i in dct:
        if year == dct[i]['Year']:
            lst.append(dct[i])
    dct1[year] = lst
    return dct1
```

```python
# create dictionary of areas to store the number of hurricanes involved in
def area_count():
    d = {}
    for i in dct:
        for area in dct[i]['Areas Affected']:
            if area not in d:
                d[area] = 1
            else:
                d[area] += 1
    return d
```

```python
area_count()
```

```python
# find most frequently affected area and the number of hurricanes involved in
def max_affected():
    m = {}
    max = 0
    n = ''
    for i in area_count():
        if area_count()[i] > max:
            max = area_count()[i]
            n = i
    m[n] = max
    return m

max_affected()
```

```python
# find highest mortality hurricane and the number of deaths
def max_deaths():
    m = {}
    max = 0
    n = ''
    for i in dct:
        if dct[i]['Deaths'] > max:
            max = dct[i]['Deaths']
            n = i
    m[n] = max
    return m
    
max_deaths()
```

```python
# categorize hurricanes in new dictionary with mortality severity as key
def mortality_rating():
    d = {0: [], 1: [], 2: [], 3: [], 4: [], 5: []}
    for i in dct:
        if dct[i]['Deaths'] == 0:
            d[0].append(dct[i])
        elif dct[i]['Deaths'] <= 100:
            d[1].append(dct[i])
        elif dct[i]['Deaths'] <= 500:
            d[2].append(dct[i])
        elif dct[i]['Deaths'] <= 1000:
            d[3].append(dct[i])
        elif dct[i]['Deaths'] <= 10000:
            d[4].append(dct[i])
        else:
            d[5].append(dct[i])
    return d

mortality_rating()
```

```python
# find highest damage inducing hurricane and its total cost
def max_damage():
    d = {}
    max = 0
    n = ''
    for i in dct:
        if dct[i]['Damage'] == 'Damages not recorded':
            continue
        elif int(dct[i]['Damage']) > max:
            max = dct[i]['Damage']
            n = i
    d[n] = max
    return d

max_damage()
```

```python
# categorize hurricanes in new dictionary with damage severity as key
def damage_scale():
    d = {0: [], 1: [], 2: [], 3: [], 4: [], 5: []}
    for i in dct:
        if dct[i]['Damage'] == 'Damages not recorded' or dct[i]['Damage'] == 0:
            d[0].append(dct[i])
        elif int(dct[i]['Damage']) <= 100000000:
            d[1].append(dct[i])
        elif int(dct[i]['Damage']) <= 1000000000:
            d[2].append(dct[i])
        elif int(dct[i]['Damage']) <= 10000000000:
            d[3].append(dct[i])
        elif int(dct[i]['Damage']) <= 50000000000:
            d[4].append(dct[i])
        else:
            d[5].append(dct[i])
    return d

damage_scale()
```
